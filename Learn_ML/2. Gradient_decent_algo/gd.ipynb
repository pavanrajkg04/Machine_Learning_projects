```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent Function
def gradient_descent(x, y):
    m_curr = b_curr = 0  # Initial slope (m) and intercept (b)
    rate = 0.01          # Learning rate (experiment with this value)
    iterations = 100     # Number of iterations
    n = len(x)           # Number of data points
    
    plt.scatter(x, y, color='red', marker='+', linewidths=5)  # Scatter plot of data points

    # Gradient Descent Loop
    for i in range(iterations):
        y_pred = m_curr * x + b_curr  # Predicted values based on current m and b
        
        # Plot the intermediate regression line
        if i % 10 == 0:  # Plot every 10 iterations to visualize progress
            plt.plot(x, y_pred, color='green', alpha=0.5, label=f'Iteration {i}' if i == 0 else "")
        
        # Calculate gradients (partial derivatives)
        md = -(2/n) * sum(x * (y - y_pred))  # Derivative with respect to m
        mb = -(2/n) * sum(y - y_pred)       # Derivative with respect to b

        # Update parameters using gradients
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * mb

        # Print progress (optional)
        print(f"Iteration {i+1}: m = {m_curr:.4f}, b = {b_curr:.4f}, Loss = {sum((y - y_pred)**2) / n:.4f}")
    
    # Final Regression Line
    y_pred_final = m_curr * x + b_curr
    plt.plot(x, y_pred_final, color='blue', linewidth=2, label='Final Regression Line')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Gradient Descent for Linear Regression')
    plt.show()

# Example Data
x = np.array([1, 2, 3, 4, 5])  # Independent variable
y = np.array([5, 7, 9, 11, 13])  # Dependent variable

# Run the Gradient Descent
gradient_descent(x, y)
```
