```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv(r"canada_per_capita_income.csv")
data.head()
```

```python
data.describe()
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(data['year'],data['per capita income (US$)'],color='red',marker='+')
```

```python
LR = linear_model.LinearRegression()
data_train = data[:-5]
LR.fit(data_train[['year']],data_train[['per capita income (US$)']])
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
LR.get_params()
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(data['year'],data['per capita income (US$)'],color='red',marker='+')
plt.scatter(data['year'][-5:], LR.predict(data[['year']][-5:]), color='blue', marker='o', label="Predicted (Last 5)")
plt.plot(data[["year"]],LR.predict(data[["year"]]),color="green")
```
